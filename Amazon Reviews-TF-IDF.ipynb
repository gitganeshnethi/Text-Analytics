```python
import warnings
warnings.filterwarnings("ignore")
import pandas as pd
import numpy as np
from nltk.corpus import stopwords
from nltk.stem.snowball import SnowballStemmer
from nltk.stem import WordNetLemmatizer
from matplotlib import pyplot as plt
from sklearn import preprocessing
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer
import nltk
nltk.download('stopwords')
```

```python
data = pd.read_excel('D:\\Un structured data\\Amazon.xlsx')


text = data[['reviews.text']]
text.head()
```

```python
data.shape
```

```python
text['Count'] = text['reviews.text'].apply(lambda x:len(str(x).split(" ")))

text.head()
```

```python
#removing everything except a to z
text['clean_text'] = text['reviews.text'].str.lower().str.replace('[^a-z ]','')

text.head()
```

```python
#removing the stopwords

stop = stopwords.words('english')
stop.extend(['rt','pmo','amp'])#Adding these words to stop because I don't want these words in my text

def sw(x):
    x = [word for word in x.split() if word not in stop]
    return " ".join(x)
text['split_text'] = text['clean_text'].apply(sw)
text.head()
```

```python
tfidf = TfidfVectorizer(ngram_range=(2,2))

tfidf.fit(text['split_text'])

```

```python
tfidf.vocabulary_.items()
```

```python
#Create a DTM

DTM = tfidf.fit_transform(text['split_text'])
DTM #unigram DTm
```

```python
df = pd.DataFrame(DTM.toarray(),columns=tfidf.get_feature_names())

df.head()
```

```python
sparse_matrix = df.sum()
final_table = pd.DataFrame(sparse_matrix).reset_index()
final_table = final_table.rename(columns={'index':'word',0:'freq'})
final_table.sort_values(by='freq',ascending=False).head(5)
```
